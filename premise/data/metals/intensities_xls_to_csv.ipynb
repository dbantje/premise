```python
import pandas as pd
```

```python
file_path = 'SI_2_Material_requirements.xlsx'
xls = pd.ExcelFile(file_path)
```

```python
sheets_to_exclude = ["Materials", "Technologies", "References"]
sheets_to_include = [sheet for sheet in xls.sheet_names if sheet not in sheets_to_exclude]
```

```python
combined_df = pd.DataFrame(columns=["metal", "unit", "year", "mean", "median", "min", "max", "origin_var"])
```

```python
for sheet in sheets_to_include:
    df = pd.read_excel(file_path, sheet_name=sheet, skiprows=2, header=None, usecols="A:G")
    df.columns = ['metal', 'unit', 'year', 'mean', 'median', 'min', 'max']
    df = df.dropna(how='all')
    df = df[df['metal'] != 'mean']
    df["origin_var"] = sheet
    
    combined_df = pd.concat([combined_df, df], ignore_index=True)
    
combined_df
```

```python
output_file = 'metals_db.csv'
combined_df.to_csv(output_file, index=False)
```
